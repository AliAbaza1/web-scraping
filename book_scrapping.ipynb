```python
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
import pandas as pd
```

```python
url = 'http://books.toscrape.com/catalogue/category/books/science_22/index.html'
response = requests.get(url)
print(response.text)
```

```python
soub = BeautifulSoup(response.text,'html.parser')
```

make this list to save the text to but it in a csv file.

```python
book_title = []
book_price = []
book_rate = []
```

find the element containing info we need.

```python
 title = soub.find_all("h3")
 price = soub.find_all("p",{"class":"price_color"})
 rate = soub.find_all('p',{'class':'star-rating'})
```

Loop over returned lists to extract needed info into other lists.

```python
for i in range(len(title)):
  book_title.append(title[i].text)
  book_price.append(price[i].text)
  book_rate.append(rate[i].attrs['class'][1])
```

Create csv file and fill it with values

```python
file_book = [book_title,book_price,book_rate]
export = zip_longest(*file_book)
with open("/content/book.csv",'w') as book:
  wr = csv.writer(book)
  wr.writerow(["book title","book price","book rate"])
  wr.writerows(export)
```

read the csv file with the pands library.

```python
pd.read_csv('/content/book.csv')
```
